```python
# 2D UMAP

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from umap import UMAP

# Step 1: Generate synthetic data
data, labels = make_blobs(n_samples=300, centers=3, n_features=50)

# Step 2: Initialize and fit UMAP
umap = UMAP(n_neighbors=4, min_dist=0.4, n_components=2)
reduced_data = umap.fit_transform(data)

# Step 3: Plotting the reduced data
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis')
plt.title('UMAP projection of the dataset')
plt.colorbar()
plt.show()
```

```python
# 3D UMAP

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_blobs
from umap import UMAP

# Step 1: Generate synthetic data
data, labels = make_blobs(n_samples=300, centers=4, n_features=50)

# Step 2: Initialize and fit UMAP for 3D projection
umap_3d = UMAP(n_neighbors=15, min_dist=0.1, n_components=3)
reduced_data_3d = umap_3d.fit_transform(data)

# Step 3: Plotting the reduced data in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(reduced_data_3d[:, 0], reduced_data_3d[:, 1], reduced_data_3d[:, 2],
                     c=labels, cmap='viridis', s=50, alpha=0.6)
legend = ax.legend(*scatter.legend_elements(), loc="upper right", title="Classes")
ax.add_artist(legend)
ax.set_title('UMAP 3D projection of the dataset')
plt.show()
```
